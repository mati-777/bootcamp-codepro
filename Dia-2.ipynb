```python
edad_ingreso = int(input('ingrese su edad'))

if edad_ingreso < 18:
    print('Acceso denegado, solo zona kids 🎈')

elif edad_ingreso <= 21: 
    print('Acceso permitido - No puede comprar bebidas alcoholicas 🎟️')

else:
    print('Acceso total - Disfruta el festival 🎉')    
```

```python
temperatura_usuario = float(input('ingrese la temperatura'))

print(f'la temperatura actual es de {temperatura_usuario}')

if temperatura_usuario >= -10 and temperatura_usuario <= 13:
    print("ro'y koko")

elif temperatura_usuario >= 14 and temperatura_usuario <= 23:
    print("Refresca'i")

elif temperatura_usuario >= 24 and temperatura_usuario <= 28:
    print('oikoite')

elif temperatura_usuario >= 29 and temperatura_usuario <= 45:
    print ('hakuroooo')

else:
    print ("Ñamano mba'e")           
```
